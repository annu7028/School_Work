```python
import random
import numpy as np

def Player1(cards):
    n = len(cards)
    #initalize DP table
    Val = [[0 for i in range(n)] for i in range(n)]

    #build table
    #k is used in order to build table in daigonal fashon
    #array will be the diagonal, and build the rest of the table from the 
    #diagonal up
    
    #the optimal score will be the top right 
    for gap in range(n): #gap ensures that everything below desired values is 0
        for j in range(gap, n):
            i = j - gap    
            
            #Val dependent on what player1 AND player2s picks
            #Player1 has many choices
            
            #a = val(i+2, j) --> P1 choose i, P2 choos i+1
            #b = val(i+1, j-1) --> P1 choose i, P2 choose j OR P1 choose j, P2 choose i
            #c = val(i,j-2) --> P1 choose j, P2 choose j-1
            
            #Player2's coices
            #only chosen(filling of variables) if it's the smaller of the two options
            if(i + 2 > j):
                a = 0
            else:
                a = Val[i+2][j]
                
            if(i + 1 > j -1):
                b = 0
            else:
                b = Val[i+1][j-1]
                
            if(i > j-2):
                c = 0
            else:
                c = Val[i][j-2]

            #if player1 picks i, then player2 will pick the min of what is left over in the array, leaving player1 with 
            #the max numbers of the array to pick from
            
            #if player1 picks j, then player2 will pick 
            Val[i][j] = min(cards[i] + max(a,b), cards[j] + max(b,c))
            
    return Val[0][n-1] #end position that gives total

def Game(cards):
    n = len(cards)
    sum = 0
    
    for i in range(n):
        sum += cards[i]
    P1Score = Player1(cards)
    
    P2Score = sum - P1Score
    print("Player 1 Score:", P1Score)
    print("Player 2 Score:", P2Score)
    
    #Testing
    if P1Score <= P2Score: #P1 did NOT loose
        print("Player 1 Did NOT loose!!")
        return(1)
    else: #player 2 won
        return(2)

cards = [4, 2, 6, 5]
print("INPUT:", cards)
Game(cards)
print("\n")

cards2 = [9, 8, 10, 9]
print("INPUT:", cards2)
Game(cards2)
print("\n")

BigEx = [35, 83, 47, 37, 90, 10, 13, 2, 96, 50, 93, 36, 64, 95]
print("LARGER EXAMPLE:", BigEx)
Game(BigEx)
print("\n")

#final test
final = random.sample(range(1, 101), 100)
print("FINAL EXAMPLE")
Game(final)
print("\n")

# A = [6, 2, 3, 4, 1, 5]
# print("B", A)
# Game(A)
# print("\n")
```

```python
player1_wins = 0
player2_wins = 0
for _ in range(100):
    test_set = random.sample(range(1, 101), 100)
    win = Game(test_set)
    if win == 1:
        player1_wins +=1
    if win == 2: 
        player2_wins += 1
print(player1_wins, player2_wins)
```
